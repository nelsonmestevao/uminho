```python
import networkx as nx
import matplotlib.pyplot as plt
import random
```

```python
G = nx.Graph()
```

```python
for i in range(10):
    G.add_node(i)
```

```python
while not nx.is_connected(G):
    [a, b] = random.sample(range(10), 2)
    G.add_edge(a, b)
```

```python
nx.draw(G, with_labels=True, font_weight='bold')
```

```python
print(nx.is_connected(G))
```
